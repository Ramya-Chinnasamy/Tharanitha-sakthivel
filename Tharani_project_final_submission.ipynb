```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file
df = pd.read_csv("all_items_expanded.csv")

# Categories in the CSV
categories = df.columns.tolist()

# Store selected items
selected_items = []

def show_menu():
    print("\nAvailable Categories:")
    for i, cat in enumerate(categories, start=1):
        print(f"{i}. {cat}")
    print("0. Finish and Show Selected Items")

def select_items():
    while True:
        show_menu()
        choice = int(input("\nSelect a category (0 to Finish): "))

        if choice == 0:
            break

        category = categories[choice - 1]
        items = df[category].dropna().tolist()

        print(f"\nAvailable items in {category}:")
        for i, item in enumerate(items, start=1):
            print(f"{i}. {item}")

        item_choice = int(input("Choose an item (0 to skip): "))
        if item_choice == 0:
            continue

        item_name = items[item_choice - 1]
        quantity = int(input("Enter number of units: "))

        # Units based on category
        if "Liquid" in category:
            unit = input("Enter unit (ml / L): ")
        elif "Soap" in category or "Hygiene" in category:
            unit = "pieces"
        else:
            unit = input("Enter unit (g / kg): ")

        selected_items.append((item_name, quantity, unit))

def show_graphs(final_items):
    items = list(final_items.keys())
    quantities = list(final_items.values())

    # 1. Bar Graph
    plt.figure(figsize=(8, 6))
    plt.bar(items, quantities)
    plt.title("Bar Graph - Grocery Items vs Quantity")
    plt.xlabel("Items")
    plt.ylabel("Quantity")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # 2. Pie Chart
    plt.figure(figsize=(8, 6))
    plt.pie(quantities, labels=items, autopct='%1.1f%%', startangle=140)
    plt.title("Pie Chart - Distribution of Grocery Quantities")
    plt.show()

    # 3. Line Graph
    plt.figure(figsize=(8, 6))
    plt.plot(items, quantities, marker='o', linestyle='-', color='green')
    plt.title("Line Graph - Grocery Items vs Quantity")
    plt.xlabel("Items")
    plt.ylabel("Quantity")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # 4. Horizontal Bar Graph
    plt.figure(figsize=(8, 6))
    plt.barh(items, quantities, color='orange')
    plt.title("Horizontal Bar Graph - Grocery Items vs Quantity")
    plt.xlabel("Quantity")
    plt.ylabel("Items")
    plt.tight_layout()
    plt.show()

# Run the program
select_items()

# Convert selected_items list  dictionary {item: total_quantity}
final_items = {}
for name, qty, unit in selected_items:
    if name in final_items:
        final_items[name] += qty
    else:
        final_items[name] = qty

print("\nFinal Shopping List:")
for name, qty, unit in selected_items:
    print(f"- {name}: {qty} {unit}")

# Show graphs
if final_items:
    show_graphs(final_items)
else:
    print("No items selected. Nothing to show.")
```
